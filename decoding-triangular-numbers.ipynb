```python
def is_triangular(n):

    # If the discriminant is a perfect square, then it's a triangular number
    discriminant = 1 + 8 * n
    return discriminant >= 0 and int(discriminant**0.5)**2 == discriminant

def decode(message_file):
    pyramid = {}

    # Read the pyramid structure from the file
    with open(message_file, 'r') as file:
        for line in file:
            parts = line.split()
            number, word = int(parts[0]), parts[1]
            pyramid[number] = word

    # Extract and sort the message words from the pyramid for triangular numbers
    message_words = [pyramid[i] for i in sorted(pyramid.keys()) if is_triangular(i)]

    # Join the message words to form the decoded message
    decoded_message = ' '.join(message_words)

    return decoded_message
```

```python
message_file = ("/content/sample_data/coding_qual_input.txt")
result = decode(message_file)
print(result)
```
